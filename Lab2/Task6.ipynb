```python
from utils.Models import AlexNet
from tensorflow.keras.optimizers import Adam
from utils.Plots import lossPlot, accuracyPlot
import numpy as np

img_w, img_h = 128, 128
img_ch = 1
n_epoch = 100
# parameters = [[8, 0.0001, 8], [16, 0.0001, 8], [8, 0.0001, 8]]
parameters = [[32, 0.00001, 2], [16, 0.00001, 2], [8, 0.00001, 2],[32, 0.0001, 2], [16, 0.0001, 2], [8, 0.0001, 2],
              [32, 0.00001, 4], [16, 0.00001, 4], [8, 0.00001, 4],[32, 0.0001, 4], [16, 0.0001, 4], [8, 0.0001, 4],
              [32, 0.00001, 8], [16, 0.00001, 8], [8, 0.00001, 8],[32, 0.0001, 8], [16, 0.0001, 8], [8, 0.0001, 8]]


batch_size_opt, LR_opt, base_opt, max_acc = 0, 0, 0, 0

for i in parameters:
    
    base, LR, batch_size = i
    
    print(f"Number of kernels: {base}, learning rate: {LR}, number of batches: {batch_size}")
    model = AlexNet(img_ch, img_w, img_h, base)
    model.compile(loss='BinaryCrossentropy', optimizer=Adam(learning_rate=LR), metrics=['binary_accuracy'])
    model_fit = model.fit(x_train, y_train, batch_size=batch_size, epochs=n_epoch, verbose=1, validation_data=[x_test,y_test])
    
    lossPlot(model_fit)
    accuracyPlot(model_fit)

    current_acc = np.max(model_fit.history["val_binary_accuracy"])
    if current_acc > max_acc and np.max(model_fit.history["binary_accuracy"]) != 1: 
        max_acc = current_acc
        base_opt = base
        LR_opt = LR
        batch_size_opt = batch_size
```

```python
from utils.Models import AlexNet
from tensorflow.keras.optimizers import Adam, SGD, RMSprop
from utils.Plots import lossPlot, accuracyPlot
import numpy as np

base_opt = 32
LR_opt = 0.00001
batch_size_opt = 8 
n_epoch = 100
img_ch = 1
optimizers = { 'Adam': Adam(learning_rate = LR_opt),
               'SGD': SGD(learning_rate = LR_opt),
               'RMSprop': RMSprop(learning_rate = LR_opt) 
             }

for opt_name,opt in optimizers.items():

    final_model = AlexNet(img_ch, img_w, img_h, base_opt)
    final_model.compile(loss='BinaryCrossentropy', optimizer = opt, metrics=['binary_accuracy'])
    final_model_fit = final_model.fit(x_train, y_train, batch_size=batch_size_opt, epochs=n_epoch, verbose=1, validation_data=[x_test,y_test])
        
    lossPlot(final_model_fit)
    accuracyPlot(final_model_fit)
    final_acc = np.max(final_model_fit.history["val_binary_accuracy"])

    print(f"Optimizer: {opt_name}")
    print(f"Accuracy = {final_acc}\n")
```
